```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
driver = webdriver.Chrome('/Users/Anjani Jayanth Roy/Downloads/chromedriver/chromedriver.exe')
time.sleep(5)
```

```python
final_list=[]
for x in range(1,11):
    driver.get('https://www.insiderbiz.in/company-list/?page='+str(x))
    time.sleep(5)
    web_page=driver.find_element(By.XPATH, "//table[@id='WebGrid']")
    for i in range(1,51):    
        lst={}
        CIN=web_page.find_element("xpath",'./tbody/tr['+str(i)+']/td[1]').text
        NAME=web_page.find_element("xpath",'./tbody/tr['+str(i)+']/td[2]').text
        ROC=web_page.find_element("xpath",'./tbody/tr['+str(i)+']/td[3]').text
        ADDRESS=web_page.find_element("xpath",'./tbody/tr['+str(i)+']/td[4]').text
        lst['CIN']=str(CIN)
        lst['NAME']=str(NAME)
        lst['ROC']=str(ROC)
        lst['ADDRESS']=str(ADDRESS)



        final_list.append(lst)
        
        #nex_page=web_page.find_element("xpath",'/html/body/section[2]/div/div/div/div/div[2]/div/div/div/div/ul/li[8]/a')
        #nex_page.click()
        #time.sleep(10)
```
